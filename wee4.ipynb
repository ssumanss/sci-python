```python
def lagrange(x, x_vals, y_vals):
  """
  Performs Lagrange interpolation on the given data points.

  Args:
    x: The point to interpolate at.
    x_vals: The x-values of the data points.
    y_vals: The y-values of the data points.

  Returns:
    The interpolated value at x.
  """

  n = len(x_vals)
  y = 0
  for i in range(n):
    prod = 1
    for j in range(n):
      if i != j:
        prod *= (x - x_vals[j]) / (x_vals[i] - x_vals[j])
    y += prod * y_vals[i]
  return y
```

```python
x_vals = [1, 2, 3, 4]
y_vals = [2, 4, 6, 8]
x = 5

print(lagrange(x, x_vals, y_vals))
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def plot_lagrange(n):
    x = np.linspace(2, 5, n)
    y = np.exp(x)

    x1 = np.linspace(2, 5, 100)
    y1=lagrange(x1, x, y)

    fig, ax = plt.subplots()
    ax.plot(x1, np.exp(x1), c='r', label=r"$\frac{1}{x}$")
    ax.plot(x1, y1, c='g', label=r"$\mathcal{P}(x)$")
    ax.scatter(x, y)
    ax.legend()

    return fig

for i in range(2, 6):
    fig = plot_lagrange(i)
plt.show()
```
