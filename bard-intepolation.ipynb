```python
import matplotlib.pyplot as plt
import numpy as np

# define the data points
x = np.array([0, 1, 2, 3, 4])
y = np.array([1, 2, 3, 4, 5])

# create the list of functions
funclist = [
    lambda x: 2*x - 1,
    lambda x: 3 - x**2,
    lambda x: 5 - 2*x,
    lambda x: x - 4
]

condlist = [
    x < 1,
    (x >= 1) & (x < 2),
    (x >= 2) & (x < 3),
    x >= 3
]

# create the spline interpolant
f = np.piecewise(x, condlist, funclist)

# plot the spline interpolant
plt.plot(x, f)

# plot the data points
plt.plot(x, y, 'o')

# set the labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Spline Interpolation')

# show the plot
plt.show()
```
